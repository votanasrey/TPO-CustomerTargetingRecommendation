```python
#feature engineering 
#aa_data = target['hectoliter', 'gross_amount', 'latitude', 'longitude']
aa_data = target.loc[:, ['hectoliter', 'gross_amount', 'grade', 'segment', 'province']]
aa_data = aa_data[aa_data['hectoliter'] > 0]
aa_data = aa_data[aa_data['gross_amount'] > 0]
```

```python
#check info
aa_data.info()
```

```python
#drop null values 
aa_data.dropna(inplace=True)
```

```python
#convert the data into the numpy array 
data_array = aa_data.values
data_array
```

```python
#Getting the list of Numerical and Categorical Variables
num_cols = aa_data._get_numeric_data().columns
print (num_cols)
cols = aa_data.columns
cat_cols = list(set(cols) - set(num_cols))
cat_cols
```

```python
# Standardizing all the numerical variables
from sklearn import preprocessing
num_features=aa_data.select_dtypes(include=[np.number]).columns
aa_data[num_features]=preprocessing.MinMaxScaler().fit_transform(aa_data[num_features])
aa_data.head()
```

```python
# Import label encoder
from sklearn import preprocessing
  
# label_encoder object knows 
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in columns 
aa_data['grade']= label_encoder.fit_transform(aa_data['grade'])
aa_data['segment']= label_encoder.fit_transform(aa_data['segment'])
aa_data['province']= label_encoder.fit_transform(aa_data['province'])
```

```python
aa_data.head()
```

```python
#Choosing optimal K value
cost = []
X = aa_data
plt.figure(figsize=(15,5))
for num_clusters in list(range(1,11)):
    kproto = KPrototypes(n_clusters=num_clusters, init='Huang', random_state=42,n_jobs=-2,max_iter=15,n_init=50) 
    kproto.fit_predict(X, categorical=[0])
    cost.append(kproto.cost_)

plt.plot(cost)
plt.xlabel('K')
plt.ylabel('cost')
plt.show
```

<h1 style="font-family:Open Sans; font-size:25px";><center>K Prototype Model Training</center></h1>

```python
# Running K-Prototype clustering
kpro_model = KPrototypes(n_clusters=3, init='Huang', verbose=0, random_state=42,max_iter=20, n_init=50,n_jobs=-2,gamma=0.15) 
clusters = kpro_model.fit_predict(X, categorical=[0])
```
